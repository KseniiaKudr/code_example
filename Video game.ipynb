```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df = pd.read_csv('all_games.csv')
```

```python
df
```

```python
df1 = pd.read_csv('dataminer (1).csv')
df1.head()
df1.info()
df1['Initial  release date']
```

```python
from datetime import datetime
from datetime import date
today = date.today()
df1['Initial  release date'] = pd.to_datetime(df1['Initial  release date'], errors='coerce')
df1['year'] = df1['Initial  release date'].dt.year
```

```python
df1
```

```python
df_y_s = df1.groupby('year')['Sales'].sum().reset_index()
df_y_s
```

```python
fig = plt.figure(figsize = (10,10))
plt.bar(df_y_s['year'], df_y_s['Sales'])
plt.title('Global sales by year')
plt.xlabel('Year')
plt.ylabel('Numbers of copies, m.')
plt.xticks(rotation=90)
plt.show()
```

```python
df2 = df1.groupby(['Developer(s)', 'year'])['Sales'].sum()
df2 = df2.reset_index()
df2['year'].max()
df3 = df2[df2['year']==2018]
df3
```

```python
fig = plt.figure(figsize = (10,10))
plt.bar(df2['Developer(s)'], df2['Sales'])
plt.title('Global sales by publisher')
plt.xlabel('Publisher')
plt.ylabel('Numbers of copies, m.')
plt.xticks(rotation=90)
plt.show()
```

```python
data = pd.read_csv('vgsales.csv')
data.head()
```

```python

df_y_s = data.groupby('Year')['Global_Sales'].sum().reset_index()
df_y_s
fig = plt.figure(figsize = (10,10))
plt.bar(df_y_s['Year'], df_y_s['Global_Sales'])
plt.title('Global sales by year')
plt.xlabel('Year')
plt.ylabel('Numbers of copies, m.')
plt.xticks(rotation=90)
plt.show()
```

```python
df_pub = data.groupby('Year')['Name'].count().reset_index()
df_pub.head()
```

```python
fig = plt.figure(figsize = (10,10))
plt.bar(df_pub['Year'],df_pub['Name'])
plt.title('A number of published games by year')
plt.xlabel('Year')
plt.ylabel('Number of games')
plt.xticks(rotation=90)
plt.show()
```
